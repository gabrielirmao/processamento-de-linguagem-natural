#### Importando bibliotecas e adicionando o texto escolhido para a atividade em uma variável.

O texto escolhido é a música "Não Deixe o Samba Morrer" de Alcione.

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk import bigrams
from nltk import trigrams
from nltk import ngrams
import matplotlib.pyplot as plt
from nltk.tokenize import RegexpTokenizer

caminho = open('/content/samba.txt').read()
print(caminho)
```

#### Fazendo download dos recursos do NLTK (Natural Language Toolkit).
- Realizando a tokenização de sentenças.
- Realizando a tokenização de palavras.
- Mostrando a quantidade de sentenças e palavras encontradas.

```python
nltk.download()
```

Realizando a tokenização de sentenças.

```python
nltk.sent_tokenize(caminho)
```

Realizando a tokenização de palavras.

```python
nltk.word_tokenize(caminho)
```

Mostrando a quantidade de sentenças e palavras encontradas.
<br><br>
O algoritmo só encontra uma sentença porque o texto é a letra de uma música, as muitas quebras de linha e versos repetidos atrapalham o tokenizador de identificar as sentenças corretamente.

```python
print(f"Quantidade de sentenças encontradas: {len(nltk.sent_tokenize(caminho))}")
print(f"Quantidade de palavras encontradas: {len(nltk.word_tokenize(caminho))}")
```

#### Mostrando a frequência das palavras em formato de gráfico de barras.

```python
palavras_pare = nltk.corpus.stopwords.words('portuguese')
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*')
tokens = tokenizer.tokenize(caminho)

nova_lista = []

for token in tokens:
  if token.lower() not in palavras_pare:
    nova_lista.append(token.lower())

frequencia_palavras = nltk.FreqDist(nova_lista)
comum = frequencia_palavras.most_common()
palavras = [item[0] for item in comum]
frequencias = [item[1] for item in comum]

plt.figure(figsize=(12, 6))
plt.bar(palavras, frequencias, color='lightgreen')
plt.title(f'Palavras mais frequentes - Não Deixe o Samba Morrer')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Mostrando a quantidade de:
- bigramas
- trigramas
- n-gramas

```python
token = nltk.word_tokenize(caminho)

bigramas = list(bigrams(token))
print(bigramas)
print(f"Quantidade de bigramas: {len(bigramas)}\n")

trigramas = list(trigrams(token))
print(trigramas)
print(f"Quantidade de trigramas: {len(trigramas)}\n")

ngramas = list(ngrams(token,4))
print(ngramas)
print(f"Quantidade de ngramas: {len(ngramas)}")
```
