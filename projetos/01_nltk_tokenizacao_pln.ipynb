# Análise de Tokens e N-gramas com NLTK

**Objetivo:**  
Realizar uma análise exploratória de texto utilizando tokenização,
remoção de stopwords e geração de n-gramas.

**Texto analisado:**  
Trechos retirados do discurso *"I Have a Dream"* – Martin Luther King Jr.

**Bibliotecas:**
- Matplotlib
- NLTK

### Importando bibliotecas e adicionando texto escolhido em uma variável.

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk import bigrams
from nltk import trigrams
from nltk import ngrams
import matplotlib.pyplot as plt
from nltk.tokenize import RegexpTokenizer
```

```python
texto = open('/content/dream.txt').read()
```

### Fazendo download dos recursos do NLTK (Natural Language Toolkit).
- Realizando a tokenização de sentenças.
- Realizando a tokenização de palavras.
- Mostrando a quantidade de sentenças e palavras encontradas.

```python
nltk.download('punkt')
nltk.download('stopwords')
```

#### Realizando a tokenização de sentenças.

```python
sentencas = nltk.sent_tokenize(texto)
print(sentencas)
```

#### Realizando a tokenização de palavras.

```python
palavras = nltk.word_tokenize(texto)
print(palavras)
```

#### Mostrando a quantidade de sentenças e palavras encontradas.

```python
print(f"Quantidade de sentenças encontradas: {len(nltk.sent_tokenize(texto))}")
print(f"Quantidade de palavras encontradas: {len(nltk.word_tokenize(texto))}")
```

### Pré-processamento do texto
Etapas aplicadas:
- Tokenização
- Conversão para minúsculas
- Remoção de stopwords
- Remoção de pontuação

```python
stop_words = nltk.corpus.stopwords.words('english')
tokens = nltk.word_tokenize(texto.lower())
tokens = [token for token in tokens if token.isalpha()]
tokens = [token for token in tokens if token not in stop_words]
print(tokens)
```

### Mostrando a frequência das palavras em formato de gráfico de barras.

```python
frequencia_palavras = nltk.FreqDist(tokens)
comuns = frequencia_palavras.most_common(10)
palavras = [item[0] for item in comuns]
frequencias = [item[1] for item in comuns]

plt.figure(figsize=(12, 6))
plt.bar(palavras, frequencias, color='lightgreen')
plt.title(f'Top 10 palavras mais frequentes')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Mostrando a quantidade de:
- bigramas
- trigramas
- n-gramas

```python
bigramas = list(bigrams(tokens))
print(bigramas)
print(f"Quantidade de bigramas: {len(bigramas)}\n")

trigramas = list(trigrams(tokens))
print(trigramas)
print(f"Quantidade de trigramas: {len(trigramas)}\n")

ngramas = list(ngrams(tokens, 4))
print(ngramas)
print(f"Quantidade de ngramas: {len(ngramas)}")
```
